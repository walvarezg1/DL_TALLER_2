```python
import pickle
import matplotlib.pyplot as plt

# Cargar historiales
with open("history_rnn.pkl", "rb") as f:
    history_rnn = pickle.load(f)

with open("history_lstm.pkl", "rb") as f:
    history_lstm = pickle.load(f)

with open("history_bilstm.pkl", "rb") as f:
    history_bilstm = pickle.load(f)
```

```python
# Definimos una función que grafique accuracy y loss para cada modelo

def plot_model_history(history, model_name):
    epochs = range(1, len(history['accuracy']) + 1)
    
    plt.figure(figsize=(12, 5))
    plt.suptitle(f"Evaluación del modelo {model_name}", fontsize=16, fontweight='bold')
    
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['accuracy'], label='Entrenamiento')
    plt.plot(epochs, history['val_accuracy'], label='Validación')
    plt.title(f'{model_name} - Accuracy')
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['loss'], label='Entrenamiento')
    plt.plot(epochs, history['val_loss'], label='Validación')
    plt.title(f'{model_name} - Loss')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
# Genera gráfico para modelo RNN

plot_model_history(history_rnn, "RNN")
```

```python
# Genera gráfico para modelo LSTM

plot_model_history(history_lstm, "LSTM")
```

```python
# Genera gráfico para modelo BiLSTM

plot_model_history(history_bilstm, "BiLSTM")
```

```python
# Gráfica comparando los tres modelos

# Epochs
epochs = range(1, len(history_rnn['accuracy']) + 1)

# Accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, history_rnn['accuracy'], label='RNN')
plt.plot(epochs, history_lstm['accuracy'], label='LSTM')
plt.plot(epochs, history_bilstm['accuracy'], label='BiLSTM')
plt.title('Precisión de entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()

# Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, history_rnn['loss'], label='RNN')
plt.plot(epochs, history_lstm['loss'], label='LSTM')
plt.plot(epochs, history_bilstm['loss'], label='BiLSTM')
plt.title('Pérdida (loss) de entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
